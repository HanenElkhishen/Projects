```python
import pandas as pd

df = pd.read_csv(r"D:\ITI BI Summer Training\3-ITI Statistics\data project\vgsales.csv")
df.head()
```

```python
#Top Publisher by Global Sales
df.groupby("Publisher")["Global_Sales"].sum().sort_values(ascending=False).head(10).plot(kind="bar")
```

```python
import matplotlib.pyplot as plt
region_sales = df[["NA_Sales", "EU_Sales", "JP_Sales", "Other_Sales"]].sum()
region_sales.plot(kind="bar", color=["blue", "orange", "green", "red"], figsize=(8,5))
plt.title("Total Video Game Sales by Region")
plt.ylabel("Total Sales (in millions)")
plt.xlabel("Region")
plt.show()
```

```python
#Descriptive Statistics
#      
df.groupby("Genre")["Global_Sales"].mean().sort_values().plot(kind="barh", figsize=(8,5))
plt.title("Average Global Sales by Genre")
plt.xlabel("Average Sales (millions)")
plt.ylabel("Genre")
plt.show()
```

```python
#      
df.groupby("Platform")["Global_Sales"].mean().sort_values(ascending=False).head(10).plot(kind="bar", figsize=(8,5))
plt.title("Top 10 Platforms by Average Global Sales")
plt.ylabel("Average Sales (millions)")
plt.xlabel("Platform")
plt.show()
```

```python
#     
df.groupby("Year")["Global_Sales"].sum().plot(kind="line", figsize=(10,5), marker="o")
plt.title("Global Video Game Sales Over the Years")
plt.xlabel("Year")
plt.ylabel("Total Sales (millions)")
plt.show()
```

```python
#  Outliers   
df["Global_Sales"].plot(kind="box", figsize=(6,6))
plt.title("Boxplot of Global Sales")
plt.ylabel("Global Sales (millions)")
plt.show()
```
